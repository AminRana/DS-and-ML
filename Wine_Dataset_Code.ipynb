```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset using the correct excel reader and local file name
dfwine = pd.read_excel('Wine.xlsx')

# Clean up column names: strip any accidental trailing/leading spaces
dfwine.columns = dfwine.columns.str.strip()

# 1. Basic Scatter plot
plt.figure()
plt.scatter(dfwine['Alcohol'], dfwine['Malic_Acid'])
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('Scatter plot on Wine dataset')
plt.show()

# 2. Scatter plot with different edge colors
plt.figure()
# Fixed: Cleaned up stray hidden characters at the end of this line
plt.scatter(dfwine['Alcohol'], dfwine['Malic_Acid'], edgecolors=['r', 'b', 'g']) 
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.title('Scatter plot on Wine dataset')
plt.show()

# 3. FacetGrid for various wine classes
sns.set_style("whitegrid")
sns.FacetGrid(dfwine, hue="Customer_Segment") \
   .map(plt.scatter, "Alcohol", "Malic_Acid") \
   .add_legend()
plt.show()

# 4. Pairplot (Choosing 4 major wine traits for clean viewing)
sns.pairplot(dfwine[['Alcohol', 'Malic_Acid', 'Ash', 'Total_Phenols', 'Customer_Segment']], hue="Customer_Segment", height=3)
plt.show()

# 5. Univariate Analysis with Boxplots
plt.figure()
sns.boxplot(x="Customer_Segment", y="Alcohol", data=dfwine)
plt.show()

# 6. Jointplots 
sns.jointplot(data=dfwine, x='Alcohol', y='Malic_Acid', kind='kde')
plt.show()

sns.set_style('darkgrid')
sns.jointplot(x='Alcohol', y='Malic_Acid', data=dfwine)
plt.show()

# 7. Overlaying contours on a jointplot
g = sns.jointplot(x='Alcohol', y='Malic_Acid', data=dfwine)
g.plot_joint(sns.kdeplot, color='gray', levels=5)
plt.show()

print(dfwine.info())

# 8. Joint plots with hue partitioning
sns.jointplot(x='Alcohol', y='Malic_Acid', hue='Customer_Segment', data=dfwine)
plt.show()

# 9. Joint plot with color palette settings
sns.jointplot(x='Alcohol', y='Malic_Acid', hue='Customer_Segment', palette='Oranges', data=dfwine)
plt.show()
```
