```python
#Step 1 – Import Libraries
#Description

#First, import the libraries that will be used for data analysis and visualization.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Step 2 – Load the Dataset
#Description

#Load the EuroMillions dataset into a pandas DataFrame.

df = pd.read_csv("EuroMillions_numbers.csv", sep=";")

print(df.head())
print(df.info())
print(df.describe())
```

```python
#Step 3 – Convert the Date Column
#Description
#Convert the Date column into a proper datetime format.

df["Date"] = pd.to_datetime(df["Date"])

print(df.dtypes)
```

```python
#Step 4 – Check for Missing Values
#Description
#Check whether the dataset contains missing values.

print(df.isnull().sum())
```

```python
#Step 5 – Statistical Analysis
#Description
#Calculate the most important statistics.

print("Average Jackpot:", df["Gain"].mean())

print("Maximum Jackpot:", df["Gain"].max())

print("Minimum Jackpot:", df["Gain"].min())

print("Average Number of Winners:", df["Winner"].mean())
```

```python
#Step 6 – Correlation Analysis
#Description
#Calculate correlations between numerical variables.

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))

plt.imshow(corr, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```

```python
#Step 7 – Jackpot Distribution
#Description
#Visualize how jackpots are distributed.

plt.figure(figsize=(8,5))

plt.hist(df["Gain"], bins=20)

plt.title("Jackpot Distribution")

plt.xlabel("Jackpot (€)")

plt.ylabel("Frequency")

plt.show()
```

```python
#Step 8 – Winners Distribution
#Description
#Visualize the number of jackpot winners.

plt.figure(figsize=(7,5))

df["Winner"].value_counts().sort_index().plot(kind="bar")

plt.title("Number of Jackpot Winners")

plt.xlabel("Number of Winners")

plt.ylabel("Frequency")

plt.show()
```

```python
#Step 9 – Jackpot Trend Over Time
#Description
#nalyze how jackpot values change over time.

plt.figure(figsize=(12,5))

plt.plot(df["Date"], df["Gain"])

plt.title("Jackpot Over Time")

plt.xlabel("Date")

plt.ylabel("Jackpot (€)")

plt.show()
```

```python
#Step 10 – Frequency of Main Numbers
#Description#
#Determine which numbers appeared most frequently.

numbers = pd.concat([
    df["N1"],
    df["N2"],
    df["N3"],
    df["N4"],
    df["N5"]
])

frequency = numbers.value_counts().sort_index()

plt.figure(figsize=(12,5))

frequency.plot(kind="bar")

plt.title("Frequency of Main Numbers")

plt.xlabel("Number")

plt.ylabel("Occurrences")

plt.show()
```

```python
#Step 11 – Lucky Stars Frequency
#Description
#Analyze how often each Lucky Star appears.

stars = pd.concat([
    df["E1"],
    df["E2"]
])

star_frequency = stars.value_counts().sort_index()

plt.figure(figsize=(8,5))

star_frequency.plot(kind="bar")

plt.title("Lucky Stars Frequency")

plt.xlabel("Lucky Star")

plt.ylabel("Occurrences")

plt.show()
```

```python
#Step 12 – Box Plot Analysis
#Description: Identify outliers in jackpot values.

plt.figure(figsize=(8,5))

plt.boxplot(df["Gain"])

plt.title("Box Plot of Jackpot Values")

plt.ylabel("Jackpot (€)")

plt.show()
```

```python
#Conclusions
#1.The dataset contains 1,318 EuroMillions lottery draws.
#2.There are no missing values, indicating that the dataset is clean and suitable for analysis.
The average jackpot provides an overview of the typical prize amount.
Most jackpots fall within a moderate range, while exceptionally large jackpots are rare.
Correlation analysis shows little or no relationship between jackpot values and the winning numbers, which is expected in a random lottery.
The distribution of winners indicates that zero or one jackpot winner is the most common outcome.
Some numbers appear more frequently than others, but these differences are consistent with normal random variation rather than a predictable pattern.
Overall, the dataset demonstrates the random nature of lottery draws while providing a good example for practicing descriptive statistics and data visualization.
```
