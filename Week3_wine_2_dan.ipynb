```python
# Step 1: Import the required library
import pandas as pd

# Load the Wine dataset from the Excel file
df = pd.read_excel("dataset_dan/Wine.xlsx")

# Display the first five rows of the dataset
print(df.head())

# Display information about the dataset
# (number of rows, columns, data types and missing values)
print(df.info())

# Display summary statistics for all numerical columns
print(df.describe())
```

```python
# Step 2: Check for missing values

# Count missing values in each column
missing_values = df.isnull().sum()

# Display the results
print("Missing values in each column:")
print(missing_values)

# Display the total number of missing values
print("\nTotal missing values:", missing_values.sum())
```

```python
# Step 3: Explore the distribution of wine classes

import matplotlib.pyplot as plt

# Count the number of wines in each class
class_counts = df["Customer_Segment"].value_counts().sort_index()

# Display the number of samples in each class
print("Number of wines in each class:")
print(class_counts)

# Create a bar chart
plt.figure(figsize=(6,4))
class_counts.plot(kind="bar", color=["steelblue", "orange", "green"])

# Add chart title and labels
plt.title("Distribution of Wine Classes")
plt.xlabel("Wine Class")
plt.ylabel("Number of Samples")

plt.show()
```

```python
# Step 4: Statistical analysis by wine class

# Calculate the mean value of each feature for every wine class
class_statistics = df.groupby("Customer_Segment").mean()

# Display the statistical summary
print("Average values for each wine class:")
print(class_statistics)
```

```python
# Step 5: Correlation analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Create the heatmap
plt.figure(figsize=(12, 8))

sns.heatmap(correlation_matrix,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=0.5)

# Add chart title
plt.title("Correlation Matrix of Wine Dataset")

# Display the heatmap
plt.show()
```

```python
# Step 6: Feature distribution analysis

import matplotlib.pyplot as plt

# Create histograms for all numerical features
df.hist(figsize=(15, 12), bins=20)

# Add a title
plt.suptitle("Distribution of Wine Features", fontsize=16)

# Improve spacing between plots
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Step 7: Box Plot Analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot for Alcohol by wine class
plt.figure(figsize=(8, 6))

sns.boxplot(
    x="Customer_Segment",
    y="Alcohol",
    data=df
)

# Add chart title and labels
plt.title("Alcohol Distribution by Wine Class")
plt.xlabel("Wine Class")
plt.ylabel("Alcohol")

# Display the plot
plt.show()
```

```python
# Step 8: Box Plot Analysis - Flavanoids

import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot for Flavanoids by wine class
plt.figure(figsize=(8, 6))

sns.boxplot(
    x="Customer_Segment",
    y="Flavanoids",
    data=df
)

# Add chart title and labels
plt.title("Flavanoids Distribution by Wine Class")
plt.xlabel("Wine Class")
plt.ylabel("Flavanoids")

# Display the plot
plt.show()
```

```python
# Step 9: Scatter Plot Analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df,
    x="Alcohol",
    y="Flavanoids",
    hue="Customer_Segment",
    palette="Set1",
    s=100
)

# Add chart title and labels
plt.title("Relationship between Alcohol and Flavanoids")
plt.xlabel("Alcohol")
plt.ylabel("Flavanoids")

# Display the legend
plt.legend(title="Wine Class")

# Show the plot
plt.show()
```

```python
# Step 10: Feature Importance Analysis

from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import pandas as pd

# Separate features and target
X = df.drop("Customer_Segment", axis=1)
y = df["Customer_Segment"]

# Train the Random Forest model
rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)

# Create a DataFrame with feature importance
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf.feature_importances_
})

# Sort features by importance
importance = importance.sort_values(
    by="Importance",
    ascending=False
)

# Display the results
print(importance)

# Plot feature importance
plt.figure(figsize=(10,6))

plt.barh(
    importance["Feature"],
    importance["Importance"]
)

plt.title("Feature Importance in Wine Classification")
plt.xlabel("Importance Score")
plt.ylabel("Features")

plt.gca().invert_yaxis()

plt.show()
```
